## Drop the original Price column since we no longer need it

```python
data_cleaned=data_cleaned.drop(['Price'],axis=1)
```

## Explore Multicollearity

```python
data_cleaned.columns.values
```

```python
# sklearn doesnt have tool to check Multicollearity, jump to sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
#define vars we want to check for Multicollinearity
variables = data_cleaned[['Mileage','Year','EngineV']]
vif=pd.DataFrame()
vif['VIF']=[variance_inflation_factor(variables.values,i)
            for i in range(variables.shape[1])]
vif["Features"]=variables.columns
```

```python
vif
```

## vif < 5: okay, drop the columns with VIF > 5

```python
data_no_mult=data_cleaned.drop(['Year'],axis=1)
data_no_mult
```

## create dummies

```python
data_dummies=pd.get_dummies(data_no_mult,drop_first=True)
data_dummies.head()
```

## checking the VIF of dummies

## rearrange: DV is the first col

```python
#check the data columns' heading:
data_dummies.columns.values
```

```python
cols = ['Log Price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessing=data_dummies[cols]
```

```python
data_preprocessing.head()
```

## Standardised data and create a regression

```python
y=data_preprocessing['Log Price']
x=data_preprocessing.drop(['Log Price'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
```

```python
data_prep_scaled=scaler.transform(x)
data_prep_scaled
```

## Splitting into Training and Testing:

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(data_prep_scaled, y, test_size=0.2, 
                                                 random_state=365)
```

```python
reg=LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat=reg.predict(x_train)
```

```python
#check if the target and prediction are represent a 45-degree line.
plt.scatter(y_train, y_hat)
plt.xlabel('Target(y_train)',fontsize=18)
plt.ylabel('Prediction(y_hat)',fontsize=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#checking if residuals (error) is normally distributed
sns.distplot(y_train - y_hat)
plt.title('Residuals PDF',fontsize=18)
```

## y_train - y_hat <0: target < predicted: over-estimated
## y_train > y_hat: under-estimated

```python
reg.score(x_train, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

## Testing

```python
y_hat_test=reg.predict(x_test)
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Target(y_test)',fontsize=18)
plt.ylabel('Prediction(y_hat_test)',fontsize=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#evaluate the perf of the data prediction, convert log into whole numbers
df_perf=pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_perf
```

```python
df_perf['Target']=np.exp(y_test)
df_perf
```

## missing values due to splitting process, sklearn tried to match indices.
## reset the index by overwriting the y_test:

```python
y_test=y_test.reset_index(drop=True)
df_perf['Target']=np.exp(y_test)
df_perf.head()
```

```python
#create column Residual=Target - Prediction
df_perf['Residuals']=df_perf['Target']-df_perf['Prediction']
sns.distplot(df_perf['Residuals'])
```

```python
df_perf['Difference%']=np.absolute(df_perf['Residuals']/df_perf['Target']*100)
df_perf
```
